```python
high_sal = list(df.loc[df['salary'] == 'high']['average_montly_hours'])
low_sal = list(df.loc[df['salary'] == 'low']['average_montly_hours'])
plt.hist(high_sal, bins=100, density=True);
plt.hist(low_sal, bins=100, density=True);
```
